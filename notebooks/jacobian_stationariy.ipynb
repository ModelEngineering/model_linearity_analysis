# ANALYSIS OF JACOBIAN STATIONARITY

```python
import src.constants as cn

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Helpers

```python
def plotCluster(n_cluster: int = 1):
    out_file = f"{n_cluster}_model_linearity_analysis_data.csv"
    data_path = os.path.join(cn.DATA_DIR, out_file)
    try:
        data_df = pd.read_csv(data_path, header=None, names=['value'], index_col=0)
    except:
        raise ValueError("No data for {n_cluster} clusters")
    ser = data_df["value"]
    ser = ser.clip(upper = 1)
    _ = plt.hist(ser, bins=100)
    plt.title(f"Total models: {len(ser)}. n_cluster: {n_cluster}")
    plt.xlabel("max across clusters of max coefficient of variation of Jacobian values within clusters (clipped at 1)")
    _ = plt.ylabel("number models")
    indices = [i for i in ser.index.values if isinstance(i, str)]
    return ser[indices]
```

# Histograms for non-contiguous clusters

```python
ser = plotCluster(n_cluster=1)
```
